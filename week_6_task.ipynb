### Renaming a column 

```python
# rename column
covid19_df_nigeria.rename(columns = {"Location" : "State"}, inplace = True)
```

```python
# calculating number of active cases and creating a column for active cases
covid19_df_nigeria["Active_Cases"] = covid19_df_nigeria["Total Confirmed Cases"] - (covid19_df_nigeria["Recovered"] + covid19_df_nigeria["Deaths"])
```

### creating a pivot table using the pandas liberary and pivot_table function
### sum of all the death and recovered cases.

```python
# creating pivot table
pivot_table_df = pd.pivot_table(covid19_df_nigeria, values = ["Total Confirmed Cases", "Deaths", "Recovered"], index = "State", aggfunc = max)
```

```python
# Finding out the recovery rate
pivot_table_df["Recovery Rate"] = pivot_table_df["Recovered"]*100/pivot_table_df["Total Confirmed Cases"]
```

```python
# Finding out the mortality rate
pivot_table_df["Mortality Rate"] = pivot_table_df["Deaths"]*100/pivot_table_df["Total Confirmed Cases"]
```

```python
# sorting the values based on confirmed cases column
pivot_table_df = pivot_table_df.sort_values(by = "Total Confirmed Cases", ascending = False)
```

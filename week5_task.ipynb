## Cleaning dataset to standard

### 1.Removing Duplicates

```python
# Removing duplicates from the entire dataset
df = df.drop_duplicates()
```

### 2. Cleaning dataset column by column

### 2a. Cleaning the Last_Name column

```python
# Removing unwanted characters using strip
df["Last_Name"] = df["Last_Name"].str.strip("123._/-")
```

```python
# displaying dataset to view changes
df
```

```python
# Replacing element in dataset
df["Last_Name"] = df["Last_Name"].str.replace("Ham_za", "Hamza")
```

```python
# displaying dataset to view changes
df
```

```python
# replacing element with empty string
df["Phone_Number"] = df["Phone_Number"].str.replace("N/a", "")
```

```python
# replacing element with empty string
df["Phone_Number"] = df["Phone_Number"].str.replace("NaN", "")
```

```python
# displaying dataset to view changes
df
```
